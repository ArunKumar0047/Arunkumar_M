```python
import numpy as np
```

###  Input Features

```python
input_features = np.array([[0,0],[0,1],[1,0],[1,1]])
print("Shape of the INPUT Features",input_features.shape)
print("Input Features\n")
print(input_features)
```

### Output Features

```python
output_features = np.array([[0,0,0,1]])
print(output_features)
print(output_features.shape)
output_features = output_features.reshape(4,1)
print("Shape of output_features Modified",output_features.shape)
print(output_features)
```

### Assigning weights 

```python
weights = np.array([[0.1],[0.2]])
print("Weights Shape",weights.shape)    #Assigning a random weights
print(weights)
```

#### Addition of Learning Rate and Bias

```python
lr = 0.05   #learning rate
bias = 0.3  # bias weight
```

#### Sigmoid Function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))     # 1/1+e^-x  rnage of values from 0~1
```

#### Derivative of Sigmoid function

```python
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

#### Training and updation of weigths

```python
# going to train for 10000 times

for i in range(10000):
    inputs = input_features
    
    #feed_forward input
    in_o = np.dot(inputs,weights) + bias
    
    #feed_forward output
    out_o = sigmoid(in_o)
    
    #Calculating error with back_propogation
    error = (out_o - output_features)     #target-output = ouput_features
    #going with formula 
    x = error.sum()
    print(x)
    
    #calculating derivative
    derror_douto = error
    dout_din = sigmoid_der(out_o)
    
    # multiplying
    deriv = derror_douto * dout_din
    
    #multiplying with 3rd derivative
    inputs = input_features.T
    deriv_final = np.dot(inputs,deriv)
    
    #updating the weights
    
    weights -= lr * deriv_final
    
    #updating bias weight
    
    for i in deriv:
        bias -= lr * i
```

#### Checking Value of weights and Bias

```python
print(weights)
print(bias)
```

#### Predicting Value

```python
inp = np.array([1,0])

#step_1
result_1 = np.dot(inp,weights) + bias
result_2 = sigmoid(result_1)
print(result_2)
```
